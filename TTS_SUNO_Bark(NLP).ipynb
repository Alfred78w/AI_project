```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio

# download and load all models
preload_models()

```

```python
# generate audio from text
text_prompt = """
     Au fil de mes diverses expériences de stage et de mes missions au sein d'entreprises de renom,j'ai eu le privilège d'acquérir un large éventail de compétences professionnelles solides qui ont grandement enrichi mon parcours.
"""
audio_array = generate_audio(text_prompt)

# save audio to disk
write_wav("bark_generation.wav", SAMPLE_RATE, audio_array)

# play text in notebook
Audio(audio_array, rate=SAMPLE_RATE)
```
