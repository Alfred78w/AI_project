```python
import os
from transformers import pipeline
import torch
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores.utils import DistanceStrategy
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from langchain.chains import RetrievalQA
from langchain_community.llms import HuggingFaceEndpoint
# from ragatouille import RAGPretrainedModel
from typing import List, Tuple

class Document:
    def __init__(self, page_content: str, metadata: dict):
        self.page_content = page_content
        self.metadata = metadata

class DocumentProcessor:
    def __init__(self):
        self.embedding_model_name = "thenlper/gte-small"
        self.embedding_model = HuggingFaceEmbeddings(
            model_name=self.embedding_model_name,
            multi_process=True,
            model_kwargs={"device": "cpu"},
            encode_kwargs={"normalize_embeddings": True}
        )
        self.reader_model_name = "HuggingFaceH4/zephyr-7b-beta"
        self.tokenizer = AutoTokenizer.from_pretrained(self.reader_model_name)
        self.reader_model = self._load_reader_model()
        # self.reranker = RAGPretrainedModel.from_pretrained("colbert-ir/colbertv2.0")
        self.knowledge_vector_database = None  # Will be initialized later

    def _load_reader_model(self):
        llm = HuggingFaceEndpoint(
        repo_id=self.reader_model_name,
        # max_length=128,
        top_k=10,
        temperature=0.2,
        # repetition_penalty=1.03,
        # add_to_git_credential=True,
        load_in_8bit=True,
        huggingfacehub_api_token="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"
             )
        return llm

    def initialize_vector_database(self, path: str):
        self.knowledge_vector_database = FAISS.load_local(path, self.embedding_model, allow_dangerous_deserialization=True)


    def answer_question(self, question: str, num_retrieved_docs: int = 30, num_docs_final: int = 5) -> Tuple[str, List[Document]]:
          relevant_docs = self.knowledge_vector_database.similarity_search(query=question, k=num_retrieved_docs)
          relevant_docs = [Document(doc.page_content, doc.metadata) for doc in relevant_docs]

          # Extract context from retrieved documents
          context = "\n".join([doc.page_content for doc in relevant_docs[:num_docs_final]])

          # Build prompt with alternating roles

          final_prompt = self.tokenizer.apply_chat_template(
            [
                {"role": "system", "content": """Using the information contained in the context,
                                                give a comprehensive answer to the question.
                                          Respond only to the question asked, response should be concise and relevant to the question.
                                        Provide the number of the source document when relevant.
                                        If the answer cannot be deduced from the context,  ask for externals sources."""},  # You can fill this with appropriate system messages
                {"role": "user", "content": f"Context:\n{context}\n---\nNow here is the question you need to answer.\n\nQuestion: {question}"}
            ],
            tokenize=False,
            add_generation_prompt=True
        )

          print("=> Generating answer...")
          answer = self.reader_model(final_prompt)
          return {"answer": answer, "relevant_docs": context}
```
